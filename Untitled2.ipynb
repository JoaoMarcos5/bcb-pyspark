```python
import os
import requests
import json
from pyspark.sql import SparkSession
from pyspark.sql import functions as F


os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-11-openjdk-amd64"
os.environ["PATH"] += os.pathsep + os.path.join(os.environ["JAVA_HOME"], "bin")

!java -version
```

```python
spark = (
    SparkSession.builder
    .appName("BCB_API_PySpark")
    .master("local[*]")
    .getOrCreate()
)

print(" Spark iniciado com sucesso!")
```

```python
url = "https://olinda.bcb.gov.br/olinda/servico/MPV_DadosAbertos/versao/v1/odata/Quantidadeetransacoesdecartoes(trimestre=@trimestre)?@trimestre='2023'&$top=100&$format=json"

resposta = requests.get(url)

if resposta.status_code == 200:
    dados = resposta.json()
    registros = dados["value"]
    print(f" Dados recebidos: {len(registros)} registros")
else:
    print(" Erro na requisição:", resposta.status_code)
```

```python
df = spark.createDataFrame(registros)
df.printSchema()
df.show(5, truncate=False)
```

```python

df_select = df.select("nomeBandeira", "qtdTransacoesNacionais", "valorTransacoesNacionais")
df_select.show(5)


```

```python
df_valor_medio = df.withColumn(
    "valor_medio_transacao",
    F.col("valorTransacoesNacionais") / F.col("qtdTransacoesNacionais")
)
```

```python
df_filtrado = df.filter(F.col("qtdTransacoesNacionais") > 1_000_000)
df_filtrado.show(5)
```

```python
df_ordenado = df.orderBy(F.col("valorTransacoesNacionais").desc())
df_ordenado.show(5)
```

```python
df_categoria = df.withColumn(
    "categoria_transacao",
    F.when(F.col("qtdTransacoesNacionais") > 1_000_000, "Alta")
     .otherwise("Baixa")
)
```

```python
df_valor_medio.show(5)
```

```python



print("Total de registros:", df.count())
```

```python
df_categoria.show(5)
```

```python
df.select(
    F.mean("qtdTransacoesNacionais").alias("média_transações"),
    F.mean("valorTransacoesNacionais").alias("média_valor")
).show()
```
